```python
import csv
import matplotlib.pyplot as plt
f = open('분기지표_20180614195140.csv')

gdp = csv.reader(f)
GDP = [] #전체 GDP데이터
piece = [] #GDP 7개씩 끊은거
deviation = 0
Deviation = [] #piece와 예측할 그래프의 유사정도
final = []
Q = 0
for row in gdp :
    if 'GDP' in row[0] : 
        for i in range(1,len(row)):
            GDP.append(float(row[i]))
for j in range(0, len(GDP)-7):
    for k in range(0,7):
        piece.append(float(GDP[j+k]))
        for l in range(0,len(piece)):
            deviation = deviation + (float(piece[l]) - GDP[-l-1])**2
        Deviation.append(float(deviation))
        piece = []
k = Deviation[0]
for i in range(0, len(Deviation)) :
    if Deviation[i] < k :
        k = Deviation[i]
        Q = i
for i in range(0, 7) : 
    piece.append(GDP[Q+i])
    final.append(GDP[len(GDP)-7+i])
piece.append(GDP[Q+7])
plt.plot(piece)
plt.plot(final)
plt.show()
print(piece[-1])
```
